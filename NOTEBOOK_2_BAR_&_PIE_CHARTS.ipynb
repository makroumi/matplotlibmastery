"Comparing Lemonade Flavors Like a Pro!"

## When to Use These Charts:  
- **Bar charts**: Compare different groups (e.g., "Which flavor sold best?")  
- **Pie charts**: Show parts of a whole (e.g., "What % of sales were Berry flavor?")  

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Data
flavors = ["Classic", "Berry", "Mint"]
sales = [120, 85, 30]  # Number of cups sold
```

## About the Data:  
- `flavors`: Categories to compare (will go on X-axis for bars)  
- `sales`: Values for each category (heights of bars/slices)  

```python
plt.bar(flavors, sales, color=["#FF5733", "#C70039", "#900C3F"])
plt.title("Lemonade Sales by Flavor")
plt.ylabel("Cups Sold")
plt.show()
```

## Breaking It Down:  
- `plt.bar(x, height)`:  
  - `x`: Categories (flavors)  
  - `height`: Values (sales)  
  - `color`: Custom colors for each bar (orange/red/dark red)  
- Always label your axes (`ylabel`) and title!  

```python
plt.barh(flavors, sales, color="skyblue", edgecolor="black", linewidth=1.2)  # Horizontal bars
plt.title("Horizontal Bar Chart")
plt.xlabel("Cups Sold")
plt.grid(axis="x", alpha=0.3)
plt.show()
```

## New Tricks:  
- `barh()`: Horizontal bars (flavors on Y-axis)  
- `edgecolor`: Outline color  
- `linewidth`: Thickness of outlines  
- `grid(axis="x")`: Only show vertical grid lines  

```python
plt.pie(sales, labels=flavors, autopct="%1.1f%%", explode=(0, 0.1, 0))
plt.title("Sales Ratio by Flavor")
plt.show()
```

## Pie Chart Secrets:  
- `autopct`: Shows percentages (e.g., "42.6%")  
- `explode`: "Pops out" a slice (here, Berry flavor)  
- Colors auto-picked unless specified  

```python
colors = ["#FFD700", "#FF6347", "#ADFF2F"]
plt.pie(sales,
        labels=flavors,
        colors=colors,
        shadow=True,
        startangle=90)
plt.legend(title="Flavors:")
plt.show()
```

## Pro Touches:  
- `shadow`: Adds 3D effect  
- `startangle`: Rotates pie (90° = starts at top)  
- `legend()`: Adds color-coded key  

## Exercise: Ice Cream Shop  
**Data:**  
- Flavors: `["Choco", "Vanilla", "Strawberry"]`  
- Sales: `[150, 200, 90]`  

**Tasks:**  
1. Create a **purple/blue/green bar chart**  
2. Make a pie chart with:  
   - Exploded strawberry slice  
   - Percentages shown  
   - Legend  

*(Need help? Next cell has answers!)*  

```python
# Bar chart
plt.bar(["Choco", "Vanilla", "Strawberry"], [150, 200, 90],
        color=["purple", "blue", "green"])
plt.title("Ice Cream Sales")
plt.ylabel("Cones Sold")
plt.show()

# Pie chart
plt.pie([150, 200, 90],
        labels=["Choco", "Vanilla", "Strawberry"],
        autopct="%1.1f%%",
        explode=(0, 0, 0.1),
        colors=["purple", "blue", "green"])
plt.legend(title="Flavors:")
plt.show()
```

## How Data Scientists Use These:  
1. Compare model accuracy across algorithms (bar charts)  
2. Show class distribution in datasets (pie charts)  

**Example:**  

```python
algorithms = ["Random Forest", "SVM", "Neural Net"]
accuracy = [0.85, 0.76, 0.92]
plt.bar(algorithms, accuracy, color="lightgreen")
plt.title("Model Comparison")
plt.ylabel("Accuracy")
plt.show()
```
