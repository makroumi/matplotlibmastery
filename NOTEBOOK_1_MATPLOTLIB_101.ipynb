## Why Draw Pictures of Numbers?  
Imagine you sold lemonade:  
- Monday: 50 cups  
- Tuesday: 62 cups  
- Wednesday: 55 cups  

Looking at **raw numbers is boring** . A **plot** helps you *see*:  
- Which day was busiest?  
- Up or down trend?  

 **Today's Goal**: Learn to draw this:

```python
import matplotlib.pyplot as plt

# Magic for Jupyter notebooks (makes plots appear below cells)
%matplotlib inline
```

## What This Code Does:
- `import matplotlib.pyplot as plt`: Grabs the drawing tools
- `%matplotlib inline`: Tells Jupyter "Show my drawings HERE"

## Data Preparation

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]  # X-axis
sales = [50, 62, 55, 73, 65]  # Y-axis
```

##  Understanding the Data:
- `days`: Goes on the **bottom** (like days on a calendar)
- `sales`: Goes on the **side** (like a ruler measuring cups sold)

## Basic Line Plot

```python
plt.plot(days, sales, color="purple", marker="o")
plt.title("Lemonade Sales This Week")
plt.xlabel("Day")
plt.ylabel("Cups Sold")
plt.show()
```

## Breaking Down the Plot:
1. `plt.plot()`: Connects dots with a line
   - `color="purple"`: Pick a crayon color
   - `marker="o"`: Draw circles at each point ●
2. `plt.title()`: Writes the top heading
3. `plt.xlabel()`/`ylabel()`: Labels the bottom/side rulers

## Customization

```python
plt.plot(days, sales, color="#FF5733", marker="D", linestyle=":", linewidth=2)
plt.grid(True, alpha=0.3)
plt.title("Fancy Lemonade Sales", fontsize=14)
plt.show()
```

## New Tricks:
- `color="#FF5733"`: Fancy color codes (try [color picker](https://htmlcolorcodes.com/))
- `marker="D"`: Diamond markers ♦ (options: `"s"`=■, `"*"`=★)
- `linestyle=":"`: Dotted line (try `"--"` for dashed)
- `grid=True`: Adds graph paper lines

## Exercise: Ice Cream Shop
**Data:**

- Sales: `[120, 85, 110]`

**Tasks:**
1. Plot with **green stars** ★
2. Make line **dashed** (`"--"`)
3. Title: "Weekly Ice Cream Sales"
4. Add grid lines

## Solution

```python
days = ["Mon", "Wed", "Fri"]
sales = [120, 85, 110]

plt.plot(days, sales, color="green", marker="*", linestyle="--")
plt.title("Weekly Ice Cream Sales")
plt.grid(True)
plt.show()
```
